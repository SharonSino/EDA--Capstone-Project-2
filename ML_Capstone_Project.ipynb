## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Merge ratings and books dataframes
df = pd.merge(ratings, books, on='ISBN')
df.head()
````

````python
# Merge ratings and books dataframes
df = pd.merge(ratings, books, on='ISBN')

# Drop unnecessary columns
df.drop(['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], axis=1, inplace=True)
df.head()
````

````python
# Check for missing values
df.isnull().sum()
````

````python
# Drop rows with missing values
df.dropna(inplace=True)

# Check for missing values again
df.isnull().sum()
````

````python
# Convert 'Year-Of-Publication' to numeric, coercing errors
df['Year-Of-Publication'] = pd.to_numeric(df['Year-Of-Publication'], errors='coerce')

# Replace 0 with NaN (assuming 0 represents missing or invalid data)
df['Year-Of-Publication'] = df['Year-Of-Publication'].replace(0, np.nan)

# Filter out years that are clearly in the future or before a reasonable publication year
current_year = pd.to_datetime('now').year
df = df[(df['Year-Of-Publication'] >= 1800) & (df['Year-Of-Publication'] <= current_year)].copy()

# Drop rows with remaining NaN values in 'Year-Of-Publication'
df.dropna(subset=['Year-Of-Publication'], inplace=True)

# Convert 'Year-Of-Publication' to integer after cleaning
df['Year-Of-Publication'] = df['Year-Of-Publication'].astype(int)
````

### What all manipulations have you done and insights you found?

Data Wrangling and Manipulations Performed:

Column Name Cleaning:

Applied clean_column_names to books, users, and ratings DataFrames. This standardizes column names to snake_case (e.g., Book-Title became book_title), ensuring consistency and easier programmatic access.
Initial Type Conversion:

Applied convert_types to all three DataFrames. This function attempts to convert columns to more appropriate data types (e.g., object to numeric where applicable), which can improve memory efficiency and facilitate numerical operations.
Handling year_of_publication in books:

Identified Invalid Entries: The year_of_publication column contained non-numeric values (e.g., 'DK Publishing Inc', 'Gallimard') and unrealistic year values (like 0 or years far in the future). These were initially coerced to NaN during numeric conversion.
Imputation and Outlier Treatment: Rows with year_of_publication as 0 or greater than 2024 (assuming current year is 2024, future years are outliers) were replaced with NaN. The missing values were then imputed using the median year_of_publication (which is robust to outliers) to maintain data integrity without discarding records.
Type Conversion: After cleaning and imputation, the column was converted to int type, as years are whole numbers.
Handling Missing age in users:

Identified Invalid Entries: The age column had NaN values and potentially unrealistic age values (like 0 or extremely high ages > 100). These were initially coerced to NaN during numeric conversion.
Imputation and Outlier Treatment: Rows with age as 0 or greater than 100 were replaced with NaN. The missing age values were then imputed using the median age from the remaining valid entries. This is a common and robust strategy for handling missing numerical data, especially when the distribution might be skewed or contain outliers.
Type Conversion: The age column was converted to int type after imputation.
Insights Found:

Improved Data Quality: These steps have significantly improved the quality of the books and users DataFrames by addressing structural issues (column names), type inconsistencies, and critical missing/invalid data points in key numerical columns.
Readiness for Analysis: The year_of_publication and age columns are now in a clean, numeric format, making them suitable for statistical analysis, visualization, and use in machine learning models.
Preservation of Data: Rather than dropping rows with missing or invalid data, imputation strategies were used to retain as much information as possible, which is crucial for large datasets.

1. Why did you pick the specific chart?

I picked a count plot (bar chart) because the book_rating variable is categorical (or ordinal, representing discrete rating values from 0 to 10). A count plot is ideal for visualizing the frequency distribution of such a variable, showing how many times each rating value appears in the dataset. This helps in understanding the overall sentiment or rating patterns.

##### 2. What is/are the insight(s) found from the chart?

From the distribution of book ratings, we can observe the following key insights:

Dominance of Implicit Ratings (Rating 0): A very large proportion of ratings are '0'. In this dataset, a rating of '0' often indicates an implicit rating, meaning the user read the book but did not explicitly provide a numerical rating. This suggests that a significant portion of user interactions are not explicitly rated.
Skew Towards Higher Explicit Ratings: Among explicit ratings (1-10), there's a general trend where higher ratings (e.g., 7, 8, 10) are more frequent than lower ratings (e.g., 1, 2, 3). This indicates that users who bother to give an explicit rating tend to rate books positively.
Relatively Fewer Low Explicit Ratings: Ratings of 1, 2, and 3 are much less common compared to higher explicit ratings, implying that truly disliked books are either rated implicitly as '0' or are simply not rated explicitly with low scores as often as liked books are rated highly.
Sparse Distribution for Explicit Ratings: The counts for explicit ratings, while present, are significantly lower than the count for implicit ratings. This highlights the challenge of working with explicit feedback in this dataset.

**Will the gained insights help creating a positive business impact?**

Yes, the gained insights can help create a positive business impact, particularly for a book recommendation system or an online bookstore.

*   **Understanding User Behavior:** Knowing the distribution of ratings, especially the high frequency of implicit ratings (0), highlights the importance of considering implicit feedback when building recommendation systems. Explicit ratings (1-10) are also valuable, and the skew towards higher ratings for explicit feedback suggests that users who actively rate tend to rate favorably.
*   **Targeted Marketing and Recommendations:** Identifying popular rating ranges (7, 8, 10) can inform marketing strategies and help in recommending books that are likely to receive higher explicit ratings.
*   **Identifying Potential Issues:** While not necessarily indicating negative growth in itself, the high number of implicit ratings (0) could suggest that users are not always prompted or encouraged to provide explicit feedback. A business could investigate ways to encourage more explicit ratings to gain richer data.

**Are there any insights that lead to negative growth? Justify with specific reason.**

The insights from this specific chart (distribution of book ratings) *by themselves* do not directly lead to or indicate negative growth. The presence of a large number of implicit ratings (0) is a characteristic of the dataset and how ratings were collected, rather than an indicator of negative user sentiment or business decline.

However, *if* a business solely relied on explicit ratings and ignored the vast amount of implicit feedback (rating 0), it could lead to:

*   **Inaccurate Recommendations:** The recommendation system would miss a large portion of user interactions, potentially leading to less relevant suggestions.
*   **Missed Opportunities:** The business might not fully understand which books are being consumed (even without explicit positive ratings), missing opportunities for engagement or targeted promotions based on reading history.

Therefore, while the insight itself isn't negative, *ignoring* it could negatively impact the business's ability to effectively understand and engage with its users.
